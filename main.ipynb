# Preprocess the Data
- We are going to drop unnecessary columns
- We are going to combine categorical information into a single column
- Remove duplicate rows
- Fixing coloumn values

```python
temp=base_songs_df
```

```python
temp.info()
```

```python
temp.head(10)
```

```python
# Dropping the Unnamed: 0 column
temp=temp.drop(columns=['Unnamed: 0'])
```

```python
#changing the duration in ms to seconds
temp['duration_ms']=temp['duration_ms']/1000
#renaming the column
temp.rename(columns={'duration_ms':'duration_s'},inplace=True)
```

```python
#removing the duplicates
temp.drop_duplicates(inplace=True)
```

```python
# combining the categorical columns into one
temp.insert(
    0, "song_info", temp['track_genre']+'/'+temp['artists']+'/'+temp['album_name']+'/'+temp['track_name']
)
```

```python
#dropping album_name, track_id
temp.drop(columns=['album_name','track_id'],inplace=True)
```
